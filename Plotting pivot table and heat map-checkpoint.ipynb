#### To create a pivot tableand  heatmap, we have to import some python libraries that are needed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Using the read_csv() function from pandas we will read the downloaded csv file as “myfile”. We parse the file name.

```python
myfile = pd.read_csv("life expectancy.txt")
print (myfile.head(20))
```

#### Next we create the pivot table also with pandas using pd.pivt_table() function. We have to specify which of the variables we want in the index and what we want as column and what will be in the cells as values

```python
ptable = pd.pivot_table(myfile, index= "year", columns= "country", values= "lifeExp" )
print (ptable.head(10))
```

###### now we have a pivot table that shows us the life expectancy for a particular year in a particular country

```python
heat_map = sns.heatmap(ptable)
plt.title("Heat map showing life expectancy for different countries ")
```
